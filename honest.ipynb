```python
def parse_remarks(s):
    l = s.split(' Digital Currency Address - ')[1:]
    address_l = []
   
    for item in l:
        d = {}
        split_list = item.split(' ',1)
        d['Token'] = split_list[0]
        d['Address'] = split_list[1].split(';',1)[0]
        address_l.append(d)
   
    return address_l
df['Addresses'] = df.Remarks.apply(parse_remarks)
df.info()
```

```python
cols = df.columns.to_list()[:4] + ['Remarks']
address_l = []
for index, row in df.iterrows():   
    for item in row['Addresses']:
        d={}
        d['Address'] = item['Address']
        d['Token'] = item['Token']
        for col in cols:
            d[col] = row[col]
        address_l.append(d)

address_df = pd.DataFrame(address_l)
        
    
    
    
    
```
